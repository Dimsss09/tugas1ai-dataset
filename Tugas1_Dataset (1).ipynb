```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["Diagnosis"] = encoder.fit_transform(df["Diagnosis"])
```

```python
X = df.drop("Diagnosis", axis=1)
Y = df["Diagnosis"]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.25, shuffle=True, random_state = 54)

from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression

models = [
          (DecisionTreeClassifier(),"Decision tree"),
          (RandomForestClassifier(),"Random Forest"),
          (XGBClassifier(),"XGboost"),
          (AdaBoostClassifier(),"AdaBoost"),
         (GradientBoostingClassifier(),"Gradient boosting"),
          (SVC(),"Support vectormachines"),
          (ExtraTreesClassifier(),"Extra trees"),
    (LogisticRegression(),"Logisitic regression")
         ]
best_accuracy = 0.0
best_model = None
best_model_name = ""
for model,name in models:
    print("Evaluating: ", name)
    cf = model
    cf.fit(X_train, Y_train)
    pred = cf.predict(X_test)
    acc = accuracy_score(pred, Y_test)
    if acc > best_accuracy:
        best_accuracy = acc
        best_model = model
        best_model_name = name
    print(name, " accuracy: ", acc)
print("best model is: ", best_model_name)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
best_pred = best_model.predict(X_test)
conf_matrix = confusion_matrix(best_pred,Y_test)
sns.heatmap(conf_matrix, annot=True)
plt.show()
```

```python
print(classification_report(best_pred, Y_test))
```

```python
from sklearn.model_selection import GridSearchCV
rf_hyperparameters = {
    'n_estimators': [100, 200, 300],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20,50],
    'min_samples_split': [2, 5, 10,20],
    'min_samples_leaf': [1, 2, 4,6]
}
rf_classifier = RandomForestClassifier()
Grid = GridSearchCV(estimator=rf_classifier,param_grid=rf_hyperparameters, n_jobs=-1, cv=5)
Grid.fit(X,Y)
```

```python
print("Best Parameters:", Grid.best_params_)
print("Best Score:", Grid.best_score_)
```
